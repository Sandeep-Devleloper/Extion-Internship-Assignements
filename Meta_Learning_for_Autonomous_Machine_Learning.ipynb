```python
import h2o
from h2o.automl import H2OAutoML

# Initialize H2O
h2o.init()
```

```python
# Load a dataset (Iris dataset for example)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
data = h2o.import_file(url)

# Specify column names
data.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
```

```python
# Set the target variable as a factor (for classification)
data['species'] = data['species'].asfactor()

# Split into train and test
train, test = data.split_frame(ratios=[.8], seed=1234)
```

```python
# Define the target and features
x = data.columns[:-1]  # Features
y = "species"  # Target

# Initialize H2OAutoML
aml = H2OAutoML(max_runtime_secs=600,  # 10-minute limit
                seed=1,
                include_algos=["GBM", "XGBoost", "GLM", "DeepLearning", "StackedEnsemble"])

# Train the AutoML model
aml.train(x=x, y=y, training_frame=train)
```

```python
# Display the leaderboard
leaderboard = aml.leaderboard
print(leaderboard)

# Get the best model
best_model = aml.leader
print(best_model)
```

```python
# Predict on the test set
predictions = best_model.predict(test)

# Evaluate model performance
performance = best_model.model_performance(test)
print(performance)
```

```python
# Save the model
model_path = h2o.save_model(model=best_model, path="/your/model/path", force=True)
print(f"Model saved to {model_path}")
```
